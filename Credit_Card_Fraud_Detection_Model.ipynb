```python
cols = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11',
       'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21',
       'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
```

```python
df['V1'].skew()
```

```python
for c in cols:
  print(c)
```

```python
for c in cols:
  sk = np.round( df[c].skew())

  if sk == 0 or sk == -0 :
    m = df[c].mean()
    s = df[c].std()
    low = m - (3*s)
    high = m + (3*s)
    df[c] = df[c].map(lambda x: low if x < low else high if x > high else x)

  else:

    q1 = df[c].quantile(0.25)
    q3 = df[c].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - (1.5*iqr)
    upper = q3 + (1.5*iqr)
    df[c] = df[c].map(lambda x: lower if x < lower else upper if x > upper else x)

print("Done With Errors")
```

# **Auto ML Model Building**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
```

```python
dt_model = DecisionTreeClassifier()
rfc_model = RandomForestClassifier()
lr_model = LogisticRegression()
knn_model = KNeighborsClassifier()
svc_model = SVC()
abc_model = AdaBoostClassifier()
```

```python
models = [dt_model,rfc_model,lr_model,knn_model,svc_model,abc_model]
```

```python
models
```

```python
for m in models:
  m.fit(X_train,y_train)
  print(m.score(X_test,y_test))
```
